## Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

## Loading dataset

```python
data=pd.read_csv(r"C:\Users\VivekSharma\Downloads\Admission_Predict.csv")
data
```

```python
data.drop('Serial No.',axis=1,inplace=True)
```

## Train Test Split

```python
data.columns
```

```python
X=data.drop('Chance of Admit ',axis=1)
y=data['Chance of Admit ']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
xtrain.shape,ytrain.shape
```

```python
ss=StandardScaler()
for i in xtrain.columns:
    xtrain.loc[:,i]=ss.fit_transform(pd.DataFrame(xtrain.loc[:,i]))
    xtest.loc[:,i]=ss.transform(pd.DataFrame(xtest.loc[:,i]))
```

## Deep Learning model

```python
model=Sequential()
model.add(Dense(units=7,activation='relu'))
model.add(Dense(units=17,activation='relu'))
model.add(Dense(units=1,activation='linear'))
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
history=model.fit(xtrain,ytrain,epochs=100,validation_split=0.2)
```

```python
pred=model.predict(xtest)
r2_score(ytest,pred)
```
